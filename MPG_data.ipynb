**Fuel Efficiency Analysis**

This project analyzes the fuel efficiency of various vehicles using the "mpg" (miles per gallon) dataset. The analysis includes data cleaning, visualization, and statistical modeling to understand the factors that affect fuel efficiency.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
```

```python
df = pd.read_csv("mpg_raw.csv")
df.head()
```

```python
plt.figure(figsize=(10,5))
sea.histplot(df,x="mpg", bins = 10 ,hue="origin",multiple="dodge",palette="coolwarm")
plt.title("mpg frequncy")
```

**Explanation** : The chart displays the frequency distribution of miles per gallon (mpg) for vehicles from different origins (USA, Japan, Europe), showing that most vehicles have an mpg between 10 and 20, with the USA having the highest frequency in this range.

```python
plt.figure(figsize=(10,4))
sea.jointplot(x="mpg",y="weight", data=df , kind= "hex",palette="coolwarm")
plt.show()
```

**Explanation** Distribution of Miles Per Gallon (mpg) by Origin:
The chart displays the distribution of miles per gallon (mpg) for vehicles from different origins (USA, Japan, Europe).
It shows that most vehicles have an mpg value ranging between 10 and 20, with the USA having the highest frequency in this range.
If we observe a cluster of dark hexagons in a specific area, it indicates that many data points have similar mpg and weight values in that region.

```python
sea.pairplot(data=df , hue="origin", palette="coolwarm")
```

**Explanation** This pair plot visualizes the relationships between multiple variables in a dataset, categorized by the 'origin' feature (with categories: 'us', 'japan', 'europe') using the 'coolwarm' color palette. Each subplot shows scatter plots for paired variables and histograms for individual variables, providing a comprehensive overview of potential correlations and distributions.

```python
plt.figure(figsize=(10,5))
sea.kdeplot(data=df , x="mpg",hue="origin" , palette="coolwarm")
plt.title("KDE plot ")
```

**Explanation** This KDE plot shows the distribution of miles per gallon (mpg) for cars from three origins (USA, Japan, Europe) using different colors. The plot highlights that cars from the USA generally have lower mpg, while cars from Japan and Europe have higher mpg values.

```python
import warnings
warnings.filterwarnings("ignore")
plt.figure(figsize=(10,5))
sea.barplot(data=df, x="origin", y="horsepower", estimator=sum, palette="coolwarm", alpha=1);
plt.title("Summition of Ogigin by Horsepower")
plt.show()
```

**Explanation** Certainly! This bar chart represents the total horsepower by origin for cars. The three bars correspond to the USA, Japan, and Europe. Notably, the USA has the highest total horsepower, followed by Europe and Japan.

```python
plt.figure(figsize=(10,5))
sea.relplot(data=df, x="horsepower", y="mpg", hue="origin", size="weight", palette="coolwarm")
plt.show()
```

**Explanation** This scatter plot graphically represents the relationship between horsepower and miles per gallon (mpg) for cars originating from the USA, Japan, and Europe. The x-axis denotes horsepower, while the y-axis indicates mpg. Data points are color-coded according to the car's origin, with their size reflecting the car's weight. The legend on the right categorizes cars by origin and includes a gradient scale for weight, ranging from 1800 to 4200 pounds. This comprehensive visualization enables an in-depth analysis of how horsepower and weight influence fuel efficiency across different regions of manufacture.



```python
plt.figure(figsize=(10,5))
sea.catplot(data=df, x="origin", y="horsepower", hue="cylinders", kind="bar", palette="coolwarm")
plt.show()
```

**Explanation** The bar plot shows horsepower distribution by origin (USA, Japan, Europe) and cylinder count. Cars from the USA generally have higher horsepower, especially those with 8 cylinders, while Japanese cars have lower horsepower, primarily with fewer cylinders. European cars have moderate horsepower, with a focus on 4 and 6 cylinders.

```python
plt.figure(figsize=(10,5))
sea.swarmplot(data=df, x="origin", y="mpg", hue="cylinders", palette="coolwarm")
plt.show()
```

**Explanation** The swarm plot illustrates the distribution of miles per gallon (mpg) for cars from the USA, Japan, and Europe, based on the number of cylinders. U.S. cars display a wide range with lower mpg, particularly those with 8 cylinders. In contrast, Japanese cars generally exhibit higher mpg, clustering around specific values. European cars also demonstrate higher mpg, primarily with 4 cylinders, highlighting a focus on fuel efficiency.

```python
plt.figure(figsize=(10,5))
sea.boxplot(data=df,x="origin" , y ="mpg" , hue = "cylinders" , palette="coolwarm")
```

**Explanation** This plot is a boxplot that shows the distribution of fuel efficiency (mpg) for cars based on their origin (USA, Japan, Europe), with the cars categorized by the number of cylinders . It can be observed that Japanese cars generally have higher fuel efficiency compared to their counterparts from the USA and Europe.

```python
plt.figure(figsize=(10,5))
sea.violinplot(data=df,x="origin" , y ="mpg" , palette="coolwarm")
plt.show()
```

**Explanation** The violin plot displays the distribution of miles per gallon (mpg) for cars from three regions: the USA, Japan, and Europe. The plot indicates that Japanese cars generally have higher mpg, with a concentration around the middle range, suggesting better fuel efficiency. European cars show a similar pattern but with slightly lower mpg. In contrast, cars from the USA tend to have lower mpg values, with a wider spread, indicating more variability in fuel efficiency.
